# I used groupby to get a better understanding of how many students each school tested and below that I pulled out the names of the schools, to decide which one to choose

# Based on the parsed out data above, I'll be focusing my analysis on Coney Island Preparatory Public Charter School - Ive created a new data set below

```python
df1=df[df['School Name'] == 'Coney Island Preparatory Public Charter School']
df1
```

# Below shows the average scores for each part of the Regents Exam, for this particular school, and below that is the total data set

```python
df1.groupby(['Regents Exam','Year'])['Mean Score'].count()
```

```python
df.groupby(['Regents Exam', 'Year'])['Mean Score'].count()
```

# The pairplot below shows the total tested at this school for each year from 2015-2017; based on these graph, the number of students tested increased each year

```python
sns.pairplot(df1, hue='Year')
```

# Below is a comparison of the number tested across the years

```python
sns.jointplot(x='Year', y='Total Tested', data=df1, kind='hex')
```

# What I attempted to do here was visualize what the total tested and mean scores looked like over the total time frame

It looks like the total number of students tested was highest in 2016, with 2017 not far behind.

```python
sns.boxplot(x='Year', y='Total Tested', data=df1, hue='Mean Score')
```
